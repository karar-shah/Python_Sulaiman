```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Apply the default theme
sns.set_theme()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(data=tips, x='tip',y='total_bill')
plt.show()
```

```python
sns.relplot(data=tips, x='tip',y='total_bill', col='sex')
plt.show()
```

```python
sns.relplot(data=tips, x='tip',y='total_bill',style='smoker')
plt.show()
```

```python
sns.relplot(data=tips, x='tip',y='total_bill',hue='smoker',style='sex')
plt.show()
```

```python
sns.relplot(data=tips, x='tip',y='total_bill',hue='smoker',style='sex', size='size')
plt.show()
```

```python
dots = sns.load_dataset("dots")
```

```python
dots
```

```python
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate")
plt.show()
```

```python
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",col="align")
plt.show()
```

```python
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",col="align",hue="choice")
plt.show()
```

```python
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate",col="align",hue="choice",style="choice",size="coherence")
plt.show()
```
