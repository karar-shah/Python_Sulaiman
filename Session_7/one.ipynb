### Plotting a Distplot Without the Histogram

#### Scatterplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the example diamonds dataset
diamonds = sns.load_dataset("diamonds")
```

##### 1

```python
sns.scatterplot(data=diamonds,x="carat", y="price")
plt.show()
```

##### 2

```python
sns.set_theme(style="whitegrid")
sns.scatterplot(data=diamonds,x="carat", y="price",hue="clarity")
plt.show()
```

##### 3
color palettes: https://seaborn.pydata.org/tutorial/color_palettes.html

```python
sns.set_theme(style="whitegrid")
sns.scatterplot(data=diamonds,x="carat", y="price",hue="clarity",
                palette="ch:r=-.2,d=.3_r",)
plt.show()
```

##### 4

```python
sns.set_theme(style="whitegrid")
sns.scatterplot(data=diamonds,x="carat", y="price",hue="clarity",
                palette="ch:r=-.2,d=.3_r",size="depth",)
plt.show()
```

##### 5

```python
sns.set_theme(style="whitegrid")
sns.scatterplot(data=diamonds,x="carat", y="price",hue="clarity",
                palette="ch:r=-.2,d=.3_r",size="depth", linewidth=0,)
plt.show()
```

##### 6

```python
sns.set_theme(style="whitegrid")
sns.scatterplot(data=diamonds,x="carat", y="price",hue="clarity",
                palette="ch:r=-.2,d=.3_r",size="depth", linewidth=0,
                sizes=(1, 8))
plt.show()
```

#### Facetting histograms by subsets of data

##### 1
theme: https://seaborn.pydata.org/generated/seaborn.set_theme.html#

```python
sns.set_theme(style="darkgrid")
df = sns.load_dataset("penguins")
sns.displot(df, x="flipper_length_mm",)
plt.show()
```

##### 2

```python
sns.displot(df, x="flipper_length_mm",col="species", row="sex",)
plt.show()
```

##### 3

```python
sns.displot(df, x="flipper_length_mm",col="species", row="sex",
            binwidth=3, height=3,)
plt.show()
```

##### 4
facet: breaking data variables up across multiple subplots, and combining those subplots into a single figure

```python
sns.displot(df, x="flipper_length_mm",col="species", row="sex",
            binwidth=3, height=3,facet_kws=dict(margin_titles=True))
plt.show()
```
